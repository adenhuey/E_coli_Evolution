```python
import numpy as np
import pandas as pd

partial_corr_df = pd.read_csv("partial_correlation_matrix.tsv", sep="\t", index_col=0)
```

```python
# Convert matrix to long format with param1, param2, partial correlation
partial_corr_long = partial_corr_df.stack().reset_index()  #stacks to multilevel indexes and then resets it so param1 is 1st column param2 second and partial corr is third
partial_corr_long.columns = ['param1', 'param2', 'partial_correlation']

# Remove self correlations (where param1 == param2)
partial_corr_long = partial_corr_long[partial_corr_long['param1'] != partial_corr_long['param2']]

# Keep only upper triangle to remove duplicates
partial_corr_long = partial_corr_long[partial_corr_long['param1'] < partial_corr_long['param2']]

# Save to tsv
partial_corr_long.to_csv('partial_correlation_long.tsv', sep='\t', index=False)
"""

**`.stack()`** — converts the matrix from wide format (rows and columns) into a long format where each cell becomes a row

**`.reset_index()`** — turns the row and column labels into regular columns

The result will look like:
"""
```
