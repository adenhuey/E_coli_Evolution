```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('significant-gene-gene-correlations.tsv', sep='\t')

# Get all unique genes
genes = pd.unique(df[['param1', 'param2']].values.ravel())

# Create empty matrix
matrix = pd.DataFrame(0.0, index=genes, columns=genes)

# Fill in correlations
for _, row in df.iterrows(): #for each row in df 
    matrix.loc[row['param1'], row['param2']] = row['Pearson'] #gets the gene pair for that row in df and finds the corresponding place in matrix and places the r there.
    matrix.loc[row['param2'], row['param1']] = row['Pearson']  # mirror

# Fill diagonal with 1s
np.fill_diagonal(matrix.values, 1.0)

print(matrix.shape)
print(matrix.head())
```

```python
matrix.to_csv('matrix_of_sig_gene_gene_correlations.tsv', sep='\t', index=False)
```
